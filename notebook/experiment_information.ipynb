```python
import pandas as pd
from gensim.summarization.summarizer import summarize
```

Загрузка текстов

```python
df = pd.read_csv('../data/model/uritexts/Динамика производства силового кабеля в России/texts.csv', index_col=0)
df
```

Выкидывает те для которых не получилось получить текст

```python
df = df.dropna(subset=['text'])
```

```python
df.url.values[9]
```

```python
for i, text in enumerate(df.text.values):
    print(i)
    try:
        print(summarize(text))
    except ValueError:
        print('input must have more than one sentence')
    print('-----------')
```

```python
def gen_sum_text(text):
    try:
        return summarize(text)
    except ValueError:
        return None
```

```python
df['gen_summary'] = df.text.apply(gen_sum_text)
```

```python
for i, text in enumerate(df.gen_summary.values):
    if text:
        if '2021' in text:
            print(text)
```
